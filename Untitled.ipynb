```python
import re 
import pandas as pd
```

```python
df = open('Amal batch 343.txt', 'r', encoding = 'utf-8')
data = df.read()
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\u202f[AP]M\s-\s'

messages = re.split(pattern, data)[1:]
dates = re.findall(pattern, data)
```

```python
# Create DataFrame
df = pd.DataFrame({
    'date': dates,
    'user-message': messages
})

# Convert to datetime
df['datetime'] = pd.to_datetime(df['date'], format='%m/%d/%y, %I:%M\u202f%p - ')

# Create formatted date string with AM/PM and replace original date
df['date'] = df['datetime'].dt.strftime('%m/%d/%y, %I:%M %p')

# Drop the datetime column to keep only 2 columns
df.drop(columns=['datetime'], inplace=True)

print(df[['user-message', 'date']].head())
print(f"DataFrame shape: {df.shape}")  # This will now show (11817, 2)
```

```python
df.shape
```

```python
print(df.columns)  # This will show: Index(['date', 'message', 'datetime'], dtype='object')
```

```python
# More robust pattern for WhatsApp format
users = []
messages = []
for message in df['user-message']:
    entry = re.split('([^:]+):\s', message, 1)  # Split on first colon only
    if len(entry) > 2:  # user name exists
        users.append(entry[1].strip())
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user-message'], inplace=True)

df
```

```python
# Convert your current date string back to datetime
temp_dates = pd.to_datetime(df['date'], format='%m/%d/%y, %I:%M %p')

# Extract year and month
df['year'] = temp_dates.dt.year
df['month_name'] = temp_dates.dt.month_name()
df['day'] = temp_dates.dt.day
df['hour'] = temp_dates.dt.hour
df['minute'] = temp_dates.dt.minute

print(df)
print(f"DataFrame shape: {df.shape}")
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words("english"))

# Function to remove stopwords
def remove_stopwords(text):
    words = text.lower().split()
    filtered = [w for w in words if w not in stop_words]
    return " ".join(filtered)

# Apply on message column
temp['message_no_stopwords'] = temp['message'].apply(remove_stopwords)
```

```python
from collections import Counter
import pandas as pd

# Combine all cleaned messages
all_words = " ".join(temp['message_no_stopwords']).split()

# Keep only words with exactly 4 alphabets
filtered_words = [w for w in all_words if len(w) >= 4]

# Top 20 most common (4-letter only)
top_20 = Counter(filtered_words).most_common(20)

# Convert to DataFrame
top_df = pd.DataFrame(top_20, columns=['word', 'count'])

top_df
```
